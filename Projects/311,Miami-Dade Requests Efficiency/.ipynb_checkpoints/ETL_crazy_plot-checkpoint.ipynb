# 311_Service_Requests__MiamiDade_County__2013

```python
#create df from csv
services13_original_df = pd.read_csv("db/311_Service_Requests__MiamiDade_County__2013.csv")
services13_original_df.head()
```

## Department 1: Citations_and_Tags

Department in charge of animals. (Accounts, tags, lost & Funds) 

```python
Citations_and_Tags13_df = services13_original_df.loc[(services13_original_df["case_owner"] == "Citations_and_Tags")]
Citations_and_Tags13_df.head()
```

```python
Citations_and_Tags13_df = Citations_and_Tags13_df.drop(columns=['ticket_id','issue_description','created_year_month','method_received','sr_xcoordinate','sr_ycoordinate','ticket_year', 'ObjectId', 'state', 'ticket_created_date_time','street_address','case_owner_description','ticket__last_update_date_time','latitude','longitude','location_city','ticket_status' ])
```

```python
Citations_and_Tags13_df.head()
```

```python
Citations_and_Tags13 = pd.DataFrame()
```

```python
Citations_and_Tags13["city"] = Citations_and_Tags13_df["city"]
Citations_and_Tags13["department"] = Citations_and_Tags13_df["case_owner"]
Citations_and_Tags13["completed days"]= Citations_and_Tags13_df["actual_completed_days"]
Citations_and_Tags13["closed_date"]= Citations_and_Tags13_df["ticket_closed_date_time"]

Citations_and_Tags13.head()
```

```python
Citations_and_Tags13['closed_date'] = Citations_and_Tags13['closed_date'].astype(str)
Citations_and_Tags13['closed_date_year'] = Citations_and_Tags13['closed_date'].str.split('-').str.get(0)
Citations_and_Tags13['closed_date_month'] = Citations_and_Tags13['closed_date'].str.split('-').str.get(1)
Citations_and_Tags13.head()
```

```python
import numpy as np
Citations_and_Tags13['closed_date_month'] = pd.to_numeric(Citations_and_Tags13['closed_date_month'], errors='coerce').fillna(0).astype(np.int64)
Citations_and_Tags13['closed_date_month'] = Citations_and_Tags13['closed_date_month'].astype(int)
```

```python
# To convert mont form int to month in string XXX (Jan, Feb, Mar, etc)
import calendar
Citations_and_Tags13['Month'] = Citations_and_Tags13['closed_date_month'].apply(lambda x: calendar.month_abbr[x])
```

```python
Citations_and_Tags13 = Citations_and_Tags13.dropna()
```

```python
Citations_and_Tags13.head()
```

```python
Citations_and_Tags13.reset_index().head()
```

```python
Citations_and_Tags13 = Citations_and_Tags13.loc[Citations_and_Tags13["completed days"] > 0, :]
```

```python
Citations_and_Tags13 = Citations_and_Tags13.groupby("Month").mean()
Citations_and_Tags13.head()
```

```python
Citations_and_Tags13.to_csv("db/temp/Citations_and_Tags13.csv", index=True, header=True)
```
