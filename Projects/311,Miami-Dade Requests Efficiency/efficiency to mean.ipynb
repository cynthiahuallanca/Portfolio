```python
from sqlalchemy import create_engine
from config import password
import pymysql
pymysql.install_as_MySQLdb()
import pandas as pd
```

```python
rds_connection_string = f"root:{password}@localhost/services_311_db"
engine = create_engine(f'mysql://{rds_connection_string}')
```

```python
#check table
zip_efficiency_df = pd.read_sql_query('select * from zip_efficiency', con=engine)
zip_efficiency_df.head(10)
```

```python
zip_df = zip_efficiency_df.groupby(['zip_code'])
zip_df = zip_df.describe()
```

```python
zip_mean = zip_df.reset_index()
zip_mean.head()
```

```python
zip_mean.head()
```

```python
zip_mean = zip_mean.drop('mean', axis=1, level=0)
```

```python
zip_mean.columns = zip_mean.columns.droplevel()
zip_mean.head()
```

```python
zip_mean.columns = ['zip_code', 'count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max']
zip_mean.head()
```

```python
zip_mean = zip_mean[['zip_code','mean']]
zip_mean.head()
```

```python
zip_mean = zip_mean.sort_values('mean')
zip_mean.head(100)
```

```python
zip_mean = zip_mean.reset_index()
zip_mean.head()
```

```python
zip_mean = zip_mean[['zip_code','mean']]
zip_mean.head()
```

```python
zip_mean.to_sql(name='zip_mean', con=engine, if_exists='replace', index=False)
```
