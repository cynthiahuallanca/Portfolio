## 1 . Cleaning Data 

```python
# Dependencies
import pandas as pd
import json 
import sys
import os
```

```python
# Read original dataset. Date range varies from Nov 2017 up to June 2018 over USA.
us_csv = "Resources/large_USvideos.csv"
original_df = pd.read_csv(us_csv)
original_df.head()
```

```python
# Identify incomplete rows
original_df.count()
```

```python
# Investigate Missing Rows:
original_df[original_df["video_id"].isna()].head(10)
```

```python
# Deleting columns with no relevant info for this specific project
del original_df['ratings_disabled']
del original_df['comments_disabled']
del original_df['video_error_or_removed']
del original_df['thumbnail_link']
del original_df['description']

# Rename column name 
df= original_df.rename(columns={"video_id": "id",
                    "trending_date": "Trending Date",
                    "title": "Title",
                    "channel_title": "Channel",
                    "category_id": "Category",
                    "publish_time": "Publish Date",
                    "tags": "Tags",
                    "views": "Views",
                    "likes": "Likes",
                    "dislikes": "Dislikes",
                    "comment_count": "Comments"})

df.head()
```

```python
# Check new counts
df.count()
```

```python
# Check data Type
df.dtypes
```

```python
# Convert column Title to propper case  
df["Title"] = df.Title.str.capitalize()
df.head()
```

```python
# Creating a DataFrame of top videos with more days as "trending" (video_count) to get the categories and add the names:

video_frequence_df = df.groupby(['id', 'Category'])['id'].agg(
                {"video_frequence": len
                }).sort_values("video_frequence", ascending=False).reset_index()
```

```python
# Category frequence on trending videos:
category_frequence = video_frequence_df.groupby("Category").id.nunique()
category_frequence.sort_values(ascending=False)
category_frequence_df = pd.DataFrame(category_frequence) 
category_frequence_df= category_frequence_df.rename(columns={'id': "category_frequence"})
category_frequence_df.sort_values('category_frequence',ascending=False).head(100)
```
